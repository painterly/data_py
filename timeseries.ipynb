```python
from pandas.io.data import DataReader
import matplotlib.pyplot as plt
import datetime
import pandas as pd
```

#### Read the data

```python
aapl = DataReader("aapl", "yahoo", datetime.datetime(2007, 1, 1), 
    datetime.datetime(2012,1,1))
```

#### Plot stock price and volume

```python
top = plt.subplot2grid((4,4), (0, 0), rowspan=3, colspan=4)
top.plot(aapl.index, aapl["Close"])
plt.title('AAPL Price from 2007 - 2012')

bottom = plt.subplot2grid((4,4), (3,0), rowspan=1, colspan=4)
bottom.bar(aapl.index, aapl['Volume'])
plt.title('AAPL Trading Volume')

plt.gcf().set_size_inches(15,8)
```

#### Calculate moving averages

```python
aapl['30_MA_Open'] = pd.stats.moments.rolling_mean(aapl['Open'], 30)
aapl['150_MA_Open'] = pd.stats.moments.rolling_mean(aapl['Open'], 150)
aapl['30_MA_Open'].tail() 
```
